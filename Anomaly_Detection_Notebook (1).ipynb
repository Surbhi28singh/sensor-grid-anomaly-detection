# ðŸ“Š Anomaly Detection with Sensor Data
This notebook demonstrates anomaly detection using scatter plots and heatmaps.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For reproducibility
np.random.seed(42)
```

```python

# Generate synthetic sensor dataset
n_samples = 200
sensor_1 = np.random.normal(loc=50, scale=10, size=n_samples)
sensor_2 = np.random.normal(loc=30, scale=5, size=n_samples)

# Add anomalies
sensor_1[::20] = sensor_1[::20] + np.random.normal(30, 5, size=n_samples//20)
sensor_2[::25] = sensor_2[::25] - np.random.normal(20, 5, size=n_samples//25)

df = pd.DataFrame({'Sensor_1': sensor_1, 'Sensor_2': sensor_2})
df.head()
```

```python

plt.figure(figsize=(8,6))
plt.scatter(df['Sensor_1'], df['Sensor_2'], c='blue', label='Normal Data')

# Highlight anomalies
anomalies = df[(df['Sensor_1'] > 80) | (df['Sensor_2'] < 15)]
plt.scatter(anomalies['Sensor_1'], anomalies['Sensor_2'], c='red', label='Anomalies')

plt.xlabel('Sensor 1')
plt.ylabel('Sensor 2')
plt.title('Scatter Plot with Anomalies')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
